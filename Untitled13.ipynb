```python
import torch
import torch.nn as nn
import torch.optim as optim
import requests


url = "https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt"
text = requests.get(url).text[:100000]
chars = sorted(list(set(text)))
vocab_size = len(chars)
char_to_idx = {ch:i for i,ch in enumerate(chars)}
idx_to_char = {i:ch for ch,i in char_to_idx.items()}


seq_len = 100
data = torch.tensor([char_to_idx[c] for c in text], dtype=torch.long)
num_seq = len(data)//(seq_len+1)
data = data[:num_seq*(seq_len+1)]
data = data.view(-1, seq_len+1)
X = data[:, :-1]
y = data[:, 1:]


class CharRNN(nn.Module):
    def __init__(self, vocab_size, emb_dim, hidden_dim, rnn_type='lstm'):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, emb_dim)
        if rnn_type == 'rnn':
            self.rnn = nn.RNN(emb_dim, hidden_dim, batch_first=True)
        elif rnn_type == 'lstm':
            self.rnn = nn.LSTM(emb_dim, hidden_dim, batch_first=True)
        elif rnn_type == 'gru':
            self.rnn = nn.GRU(emb_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, x, state=None):
        emb = self.embed(x)
        out, state = self.rnn(emb, state)
        out = self.fc(out)
        return out, state


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def train_model(model_type):
    model = CharRNN(vocab_size, 64, 128, model_type).to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()
    for epoch in range(20):
        model.train()
        total_loss = 0
        for i in range(0, X.size(0), 32):
            batch_X = X[i:i+32].to(device)
            batch_y = y[i:i+32].to(device)
            out, _ = model(batch_X)
            loss = criterion(out.transpose(1,2), batch_y)
            optimizer.zero_grad()
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 5)
            optimizer.step()
            total_loss += loss.item()
        print(f"{model_type} epoch {epoch+1} loss {total_loss:.2f}")
    return model


lstm = train_model('lstm')
```

```python
#  Seq2Seq EN-FR (dataset Tatoeba tiny)
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import random


url = "https://www.manythings.org/anki/fra-eng.zip"
import zipfile, io, requests
r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
pairs = []
with z.open('fra.txt', 'r') as f:
    for line in f:
        try:
            line = line.decode('utf-8').strip()
            eng, fra, _ = line.split('\t')
            pairs.append((eng, fra))
        except:
            pass
pairs = pairs[:10000]


def tokenize(s):
    return s.lower().split()

SOS, EOS, PAD, UNK = '<sos>', '<eos>', '<pad>', '<unk>'
en_vocab = {PAD:0, SOS:1, EOS:2, UNK:3}
fr_vocab = {PAD:0, SOS:1, EOS:2, UNK:3}

for eng, fra in pairs:
    for tok in tokenize(eng):
        if tok not in en_vocab: en_vocab[tok]=len(en_vocab)
    for tok in tokenize(fra):
        if tok not in fr_vocab: fr_vocab[tok]=len(fr_vocab)

en_idx2word = {v:k for k,v in en_vocab.items()}
fr_idx2word = {v:k for k,v in fr_vocab.items()}

def encode(s, vocab):
    return [vocab.get(tok, vocab[UNK]) for tok in tokenize(s)]

max_len = 20
class TransDataset(Dataset):
    def __init__(self, pairs):
        self.pairs = pairs
    def __len__(self): return len(self.pairs)
    def __getitem__(self, idx):
        eng, fra = self.pairs[idx]
        src = [en_vocab[SOS]] + encode(eng, en_vocab) + [en_vocab[EOS]]
        tgt = [fr_vocab[SOS]] + encode(fra, fr_vocab) + [fr_vocab[EOS]]
        return torch.tensor(src[:max_len]), torch.tensor(tgt[:max_len])

def collate_fn(batch):
    src, tgt = zip(*batch)
    src_len = [len(s) for s in src]
    tgt_len = [len(t) for t in tgt]
    src_padded = nn.utils.rnn.pad_sequence(src, batch_first=True, padding_value=en_vocab[PAD])
    tgt_padded = nn.utils.rnn.pad_sequence(tgt, batch_first=True, padding_value=fr_vocab[PAD])
    return src_padded, tgt_padded, torch.tensor(src_len), torch.tensor(tgt_len)

dataset = TransDataset(pairs)
loader = DataLoader(dataset, batch_size=64, shuffle=True, collate_fn=collate_fn)

class Encoder(nn.Module):
    def __init__(self, vocab_size, emb_dim, hidden_dim):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        self.lstm = nn.LSTM(emb_dim, hidden_dim, batch_first=True)
    def forward(self, src):
        embedded = self.embed(src)
        outputs, (hidden, cell) = self.lstm(embedded)
        return hidden, cell

class Decoder(nn.Module):
    def __init__(self, vocab_size, emb_dim, hidden_dim):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        self.lstm = nn.LSTM(emb_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)
    def forward(self, tgt, hidden, cell):
        embedded = self.embed(tgt)
        output, (hidden, cell) = self.lstm(embedded, (hidden, cell))
        pred = self.fc(output)
        return pred, hidden, cell

class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
    def forward(self, src, tgt, teacher_forcing_ratio=0.5):
        batch_size, tgt_len = tgt.shape
        vocab_size = self.decoder.fc.out_features
        outputs = torch.zeros(batch_size, tgt_len, vocab_size).to(src.device)
        hidden, cell = self.encoder(src)
        dec_input = tgt[:,0:1]  # <sos>
        for t in range(1, tgt_len):
            out, hidden, cell = self.decoder(dec_input, hidden, cell)
            outputs[:,t,:] = out.squeeze(1)
            teacher_force = random.random() < teacher_forcing_ratio
            top1 = out.argmax(-1)
            dec_input = tgt[:,t:t+1] if teacher_force else top1
        return outputs


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
enc = Encoder(len(en_vocab), 128, 256)
dec = Decoder(len(fr_vocab), 128, 256)
model = Seq2Seq(enc, dec).to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss(ignore_index=fr_vocab[PAD])

for epoch in range(10):
    model.train()
    total_loss = 0
    for src, tgt, _, _ in loader:
        src, tgt = src.to(device), tgt.to(device)
        out = model(src, tgt)
        loss = criterion(out[:,1:,:].reshape(-1, len(fr_vocab)), tgt[:,1:].reshape(-1))
        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1)
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1} loss {total_loss:.2f}")


def greedy_decode(model, src, max_len=20):
    model.eval()
    with torch.no_grad():
        hidden, cell = model.encoder(src)
        dec_input = torch.tensor([[fr_vocab[SOS]]], device=src.device)
        outputs = []
        for _ in range(max_len):
            out, hidden, cell = model.decoder(dec_input, hidden, cell)
            pred = out.argmax(-1).item()
            if pred == fr_vocab[EOS]:
                break
            outputs.append(pred)
            dec_input = torch.tensor([[pred]], device=src.device)
    return [fr_idx2word[idx] for idx in outputs]


test_src = torch.tensor([encode("i love you", en_vocab)]).to(device)
print(greedy_decode(model, test_src))


```
